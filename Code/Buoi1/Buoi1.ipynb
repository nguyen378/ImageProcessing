### Nâng cao chất lượng ảnh

```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

Điều chỉnh độ sáng và tương phản

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    brightness = 10
    constrast = 2.3
    img2 = cv2.addWeighted(img1, constrast, np.zeros(
        img1.shape, img1.dtype), 0, brightness)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    alpha = 1.5
    beta = 50
    img2 = cv2.convertScaleAbs(img1, alpha=alpha, beta=beta)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

Làm sắc nét ảnh (Sharpening image)

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    img2 = cv2.filter2D(img1, -1, kernel)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img2 = cv2.Laplacian(img1, cv2.CV_64F)
    img2_normalized = img2 / 255.0
    img2_normalized = np.clip(img2_normalized, 0, 1)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2_normalized)
    plt.show()
```

Loại bỏ nhiễu (Removed noise)

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img2 = cv2.medianBlur(img1, 3)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img2 = cv2.GaussianBlur(img1, (5, 5), 0)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

Tăng cường màu sắc

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img0 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img1 = cv2.cvtColor(img0, cv2.COLOR_RGB2HSV)

    img1[:, :, 0] = img1[:, :, 0] * 0.7
    img1[:, :, 1] = img1[:, :, 1] * 1.5
    img1[:, :, 2] = img1[:, :, 2] * 0.5

    img2 = cv2.cvtColor(img1,cv2.COLOR_HLS2RGB)
    plt.subplot(1, 2, 1)
    plt.imshow(img0)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

thay đổi kích thước ảnh

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img2 = cv2.resize(img1,(512,512))

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

Biến đổi nghịch đảo

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img2 = cv2.bitwise_not(img1)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.show()
```

Cân bằng biểu đồ xám

```python
dir = 'img'

for img_path in os.listdir(dir):
    img_fullpath = f'{dir}/{img_path}'

    img = cv2.imread(img_fullpath)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    img2 = cv2.equalizeHist(img1)

    plt.subplot(1, 2, 1)
    plt.imshow(img1, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.imshow(img2, cmap='gray')
    plt.show()
```
